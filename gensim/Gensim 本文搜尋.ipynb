```python
import os
import json
import jieba
import re
import codecs
import logging
import pandas as pd

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

logging.info('先叫出所有檔案路徑')
files = ['{}/{}'.format(folder,file) 
         for folder,folders,files in os.walk('qa_data')
         for file in files if file.endswith('.txt')]
logging.info('第一個檔案名稱:{}'.format(files[0]))

logging.info('開始製作目錄')
questions = []
for f in files:
    with open(f,'r') as qfile:
        qalist = json.load(qfile)        
        questions.extend([{'file':f,'question':qa['question'],'id':qa['id']} 
                          for qa in qalist 
                          if 'question' in qa])
        qfile.close()
        
logging.info('讀出所有停用詞')
stop_words = 'jieba_dict/stopwords.txt'
stopwords = codecs.open(stop_words,'r',encoding='utf8').readlines()
stopwords = [ w.strip() for w in stopwords ]

logging.info('開始分詞')
for q in questions:
    q['words'] = [word for word in jieba.cut(q['question']) 
                  if word not in stopwords and not re.match(string= word,pattern='^[\s]*$')]
    
logging.info('完成製作目錄')
logging.info('第一筆目錄資料:{}'.format(questions[0]))

pd.DataFrame(questions,columns=['id','question','words','file'])
```

```python
from gensim import corpora
from gensim.summarization import bm25

logging.info('開始載入bm25 model')
corpus = [d['words'] for d in questions];
bm25Model = bm25.BM25(corpus)

average_idf = sum(map(lambda k: float(bm25Model.idf[k]), bm25Model.idf.keys())) / len(bm25Model.idf.keys())
logging.info('完成載入bm25 model')

query_str = '台中有什麼好玩的嗎?'
query = list(jieba.cut(query_str))
for word in query_str.strip().split():
    query.append(word)
    
scores = bm25Model.get_scores(query,average_idf)

for idx in range(0,len(questions)):
    questions[idx]['score'] = scores[idx]
    
pd.DataFrame(sorted(questions,key = lambda q:q['score'],reverse=True),columns=['id','question','words','score'])
```
